```
!pip install yfinance
import yfinance as yf
import pandas as pd
tickers_list = ['MA', 'SAP', 'ROG.SW', '^GSPC']
data1 = yf.download(tickers_list, '2019-02-01')['Close']
data1.head()
```

```
data1.dropna(inplace=True)
```

```
data1.columns = ['MASTERCARD', 'ROCHE', 'SAP', 'S&P 500']
data_a = ((data1.pct_change()+1).cumprod())
data_a.fillna(1, inplace = True)
data_a.head()
```

```
x = data_a.index
spx = (data_a['S&P 500'] ) * 100
ma = data_a['MASTERCARD'] * 100
roche = data_a['ROCHE'] * 100
sap = data_a['SAP'] * 100
columns = data_a.columns[3:]
labs = columns.tolist()
```

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(12, 5))
ax.stackplot(x, spx, labels = labs, colors = 'lightskyblue', alpha = 0.8)
ax.plot(ma, label = 'MASTERCARD', color = 'chocolate')
ax.plot(roche, label = 'ROCHE', color = 'darkgreen')
ax.plot(sap, label = 'SAP', color = 'goldenrod')
ax.set(ylim=[80, 170])
ax.legend(fontsize=14, ncol=2)
plt.ylabel('Нормализованная динамика котировок c 02/2019, %', fontsize = 10)
plt.grid(which = 'major', color = 'grey', linestyle = '-.', linewidth = 0.5)
plt.show()
```

```
!pip install yfinance
import yfinance as yf
tickers_list = ['NKE', 'PEP', 'ROG.SW', '^GSPC']
data1 = yf.download(tickers_list, '2020-02-12')['Close']
data1.head()
```

```
data1.dropna(inplace=True)
```

```
data1.columns = ['NIKE', 'PEPSI', 'ROCHE', 'S&P 500']
data_a = ((data1.pct_change()+1).cumprod())
data_a.fillna(1, inplace = True)
data_a.head()
```

```
x = data_a.index
spx = (data_a['S&P 500'] ) * 100
nke = data_a['NIKE'] * 100
pep = data_a['PEPSI'] * 100
rog = data_a['ROCHE'] * 100
columns = data_a.columns[3:]
labs = columns.tolist()
```

```
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(8, 6))
ax.stackplot(x, spx, labels = labs, colors = 'lightskyblue', alpha = 0.8)
ax.plot(nke, label = 'NIKE', color = 'chocolate')
ax.plot(pep, label = 'PEPSI', color = 'darkgreen')
ax.plot(rog, label = 'ROCHE', color = 'goldenrod')
ax.set(ylim=[60, 110])
ax.legend(fontsize=14, ncol=2)
plt.ylabel('Нормализованная динамика котировок c 02/2020, %', fontsize = 12)
plt.xticks(rotation=0, fontsize = 8)
plt.grid(which = 'major', color = 'grey', linestyle = '-.', linewidth = 0.5)
plt.show()
```
